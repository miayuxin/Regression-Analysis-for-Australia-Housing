## Identify the normality of sales price

```
# Histogram and normal probability plot.
plt.rcParams['figure.figsize'] = (10, 8)
ax = sns.distplot(my_data['SalePrice'], fit = stats.norm)

# Get the fitted parameters used by the function.
mu, sigma = stats.norm.fit(my_data['SalePrice'])
print('\n mu = {:.2f} and sigma = {:.2f} \n'.format(mu, sigma))

# Plot the distribution.
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

# Set x tick.
fmt = '${x:,.0f}'
tick = ticker.StrMethodFormatter(fmt)
ax.xaxis.set_major_formatter(tick)

# Make the QQ-plot.
fig = plt.figure()
res = stats.probplot(my_data['SalePrice'], plot=plt)
```

Seems like 'SalePrice' is not normal. It has positive skewness but does not follow the diagonal line. 

I searched on Google that says **the log transformation is usually used to normalize our target value.** Thus, I would like to try it. 

```
# Use the numpy fuction log1p which applies log(1+x) to all elements of the column.
plt.rcParams['figure.figsize'] = (10, 8)
my_data['SalePrice'] = np.log1p(my_data['SalePrice'])

# Check the new distribution and set x tick.
ax = sns.distplot(my_data['SalePrice'], fit=stats.norm)
fmt = '${x:,.0f}'
tick = ticker.StrMethodFormatter(fmt)
ax.xaxis.set_major_formatter(tick)

# Get the fitted parameters used by the function.
mu_sigma = stats.norm.fit(my_data['SalePrice'])
print('\n mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))

# Plot the distribution.
plt.ylabel('Frequency')
plt.title('SalePrice distribution')

# QQ-plot.
fig = plt.figure()
res = stats.probplot(my_data['SalePrice'], plot=plt)
```

## Deal with missing value

```
# Counts all missing values and its percentage.
total = my_data.isnull().sum().sort_values(ascending=False)
percent = (my_data.isnull().sum() / my_data.isnull().count()).sort_values(ascending=False)
frames = [total, percent]
# pd.concat(): axis=0 means index, axis=1 means column, keys = name of the column.
missing_data = pd.concat(frames, axis=1, keys = ['Total', 'Percent'])
missing_data.head(20)
```

Since the variables of 'PoolQC', 'MiscFeature', 'Alley', and 'Fence' have missing values exceed 80%. Those variables can be considered invalid and can be removed.

```
# Remove the entire column with missing values. 
data1 = my_data.drop(missing_data[missing_data['Total']>1].index, axis=1)
# Since the variable of 'Electrical' has only 1 missing value, I can just delete the row.
data2 = data1.drop(data1.loc[data1['Electrical'].isnull()].index)
# Check the number of missing values.
data2.isnull().sum().max()
```

# Build a Linear Regression Model

Get 3:1 training - test data ratio (which means test_size = 0.25 below).

In order to avoid overfitting, I need to split data into training data and testing data. 

Overfitting means my training loss is much lower than testing loss.  

```
X = data2[['OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'TotRmsAbvGrd', 'YearBuilt']]
Y = data2['SalePrice']
print(X)
 
# Split the data into testing data and training data. 
# Already from sklearn.model_selection import train_test_split.
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=0)

X_train = np.asmatrix(X_train)
X_test = np.asmatrix(X_test)
y_train = np.ravel(y_train)
y_test = np.ravel(y_test)
```

```
# Already from sklearn.linear_model import LinearRegression.
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

**We now have a working Linear Regression model!**

## Caculate R squared

```
y_pred = regressor.predict(X_test)
print('Linear Regression R squared: %.4f' % regressor.score(X_test, y_test))
```

**Linear Regression R squared: 0.8414**

**So in my model, 84.14% of the variability in Y can be explained using X, which means it can interpret 84% of home prices.**


## Calculate root-mean-square error (RMSE)

```
# Already from sklearn.metrics import mean_squared_error.
lin_mse = mean_squared_error(y_pred, y_test)
lin_rmse = np.sqrt(lin_mse)
print('Linear Regression RMSE: %.4f' % lin_rmse)
```

**The standard deviation of the residuals (prediction errors) is 0.1565, seems like my model is not bad.**
